### 📌 프로젝트 소개

- 제품 및 공장의 불량률을 나타내는 제조 데이터는 민감한 정보이므로 기업이 공개적으로 공유하지 않는 경우가 많습니다.
- 그럼에도 불구하고 제조 공정의 불량률을 줄이거나 개선하는 데 도움이 될 수 있으므로 중요한 데이터입니다.
- 실제로 현장에서의 목표는 일반적인 데이터 분석처럼 단순히 고성능 예측 모델을 개발하는 것이 아닙니다. 대신 `어떤 요인이 결함에 영향을 미치는지 이해하고 이를 줄이기 위한 효과적인 전략을 결정`하는 데 중점을 둡니다.
- 따라서 데이터 분석가의 임무는 `다양한 데이터를 분석하고 제품 불량 발생률을 줄일 수 있는 솔루션을 제공`하는 것입니다.


### ✅ 데이터 안내
- 본 실습에 사용된 데이터는 실제 제조과정에서 나온 데이터로 익명처리되어 어느 회사의 제품인지 식별이 불가능합니다.
- 제조업 데이터를 다루다 보면 제조업 분야에서 흔히 발생하는 `데이터 불균형`을 자주 접하게 됩니다.
- 비식별화된 제조 데이터에서 `종속변수 Y`는 `제품의 불량 여부`를 나타냅니다.
- X로 표시된 나머지 변수는 독립변수입니다.

```python
import pandas as pd
# 데이터 로드
df = pd.read_csv('1주_실습데이터.csv')
columns_to_check = ['X1','X2','X3','X5','X6','X7','X8','X9','X10','X11','X12','X14','X15','X16','X17']

# 각 변수별 이상치 제거
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)  # 1사분위수
    Q3 = df[column].quantile(0.75)  # 3사분위수
    IQR = Q3 - Q1  # IQR 계산
    lower_bound = Q1 - 1.5 * IQR  # 이상치 하한선
    upper_bound = Q3 + 1.5 * IQR  # 이상치 상한선

    # 이상치 제거
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 이상치 제거 후의 데이터프레임 출력
print(df)
```

# 4.이상치 제거

```python
bounds = {}

for column in data.columns:
    lower_bound = data[column].quantile(0.01)
    upper_bound = data[column].quantile(0.99)
    bounds[column] = (lower_bound, upper_bound)

# 모든 컬럼에 대한 조건을 동시에 적용하여 이상치 제거
for column in data.columns:
    data = data[(data[column] >= bounds[column][0]) & (data[column] <= bounds[column][1])]

# 이상치가 제거된 데이터 확인
data.columns
```

# 5.t-test 검정

```python
import pandas as pd
import numpy as np
import scipy
## T-test 검정
t_test = []

for idx, col in enumerate(data.columns):
    t = scipy.stats.ttest_ind(data[data['Y']==0][col],
                              data[data['Y']==1][col])
    t_test.append([col, t[0], t[1]])

t_test_data = pd.DataFrame(t_test, columns=['col', 't', 'p-value'])

## P-value가 0.05보다 작은 변수만 추출
t_test_data = t_test_data[t_test_data['p-value']<0.05]
t_test_data
```

# 6. 모델훈련

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
import pandas as pd
import matplotlib.pyplot as plt
# 특성과 타겟 변수 분리
X = data.drop('Y', axis=1)
y = data['Y']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# XGBoost 모델 훈련
model = xgb.XGBClassifier(objective='binary:logistic', random_state=42)
model.fit(X_train, y_train)

# 변수 중요도 추출
feature_importances = model.feature_importances_
```

```python
import matplotlib.pyplot as plt
import pandas as pd
features = X_train.columns

# 가정: feature_importances는 이미 계산된 피처 중요도 배열
# features = X_train.columns # X_train이 pandas DataFrame일 경우
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# 막대 그래프로 시각화
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # y축 순서 뒤집기

# x축 눈금 간격 설정
plt.xticks(np.arange(min(importance_df['Importance']), max(importance_df['Importance']) + 0.1, 0.1))

plt.show()
```

```python
# 데이터프레임에서 'x3' 열을 기준으로 10개의 구간으로 분할합니다. 이 때 구간의 경계는 'cut_points'에 저장됩니다.
cut_points = pd.cut(data['X3'], bins=10)

# 각 구간별로 불량 여부를 확인하여 계산합니다.
# 'groupby'를 사용하여 각 구간에 대한 불량 여부의 평균을 계산합니다.
defect_rate_by_bin = data.groupby(cut_points)['Y'].mean()

# 각 구간별 불량률 출력
print(defect_rate_by_bin)

# 각 구간에서의 불량 발생 비율을 기준으로 불량 여부를 판단합니다.
threshold = 0.1  # 불량 발생 비율의 임계값
defective_prediction = defect_rate_by_bin > threshold

# 각 구간에서의 불량 여부를 출력합니다.
print(defective_prediction)
```

```python
# X3를 5개의 구간으로 나누기
data['X3_bin'] = pd.cut(data['X3'], bins=5, labels=['Bin1', 'Bin2', 'Bin3', 'Bin4', 'Bin5'])

# 각 구간에서의 불량률 계산
data['Defect'] = data['Y'] == 1  # 불량품 여부를 나타내는 새로운 열 추가
grouped_data = data.groupby('X3_bin')['Defect'].mean().reset_index()
grouped_data.columns = ['X3_Bin', 'Defect_Rate']
```

```python
# 막대 그래프로 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='X3_Bin', y='Defect_Rate', data=grouped_data, palette='coolwarm')
plt.title('Defect Rate by X3 Bins')
plt.xlabel('X3 Bins')
plt.ylabel('Defect Rate')
plt.show()
```

```python
# X3를 10개의 구간으로 나누기
data['X3_bin'] = pd.cut(data['X3'], bins=10, labels=[f'Bin{i+1}' for i in range(10)])

# 각 구간에서의 불량률 계산
data['Defect'] = data['Y'] == 1  # 불량품 여부를 나타내는 새로운 열 추가
grouped_data = data.groupby('X3_bin')['Defect'].mean().reset_index()
grouped_data.columns = ['X3_Bin', 'Defect_Rate']

# 막대 그래프로 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='X3_Bin', y='Defect_Rate', data=grouped_data, palette='coolwarm')
plt.title('Defect Rate by X3 Bins')
plt.xlabel('X3 Bins')
plt.ylabel('Defect Rate')
plt.show()
```

```python
# Bin1에서의 데이터 필터링
bin1_data = data[data['X3_bin'] == 'Bin1']

# Bin1을 더 작은 구간으로 나누기
bin1_data['X3_sub_bin'] = pd.cut(bin1_data['X3'], bins=5, labels=[f'Bin1_Sub{i+1}' for i in range(5)])

# 각 작은 구간에서의 불량률 계산
bin1_defect_rates = bin1_data.groupby('X3_sub_bin')['Defect'].mean().reset_index()
bin1_defect_rates.columns = ['X3_Sub_Bin', 'Defect_Rate']

# 막대 그래프로 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='X3_Sub_Bin', y='Defect_Rate', data=bin1_defect_rates, palette='coolwarm')
plt.title('Defect Rate by X3 Sub-Bins in Bin1')
plt.xlabel('X3 Sub-Bins')
plt.ylabel('Defect Rate')
plt.show()
```

```python
# Bin1_Sub3 데이터 필터링
bin1_sub3_data = bin1_data[bin1_data['X3_sub_bin'] == 'Bin1_Sub3'].copy()

# Bin1_Sub3을 더 작은 구간으로 나누기
bin1_sub3_data['X3_sub_sub_bin'] = pd.cut(bin1_sub3_data['X3'], bins=5, labels=[f'Bin1_Sub3_Sub{i+1}' for i in range(5)])

# 각 하위 구간의 데이터 포인트 수 계산
data_counts = bin1_sub3_data['X3_sub_sub_bin'].value_counts()

# 결과 출력
print(data_counts)
```

```python
# Bin1_Sub3_Sub3에서의 불량률 계산
bin1_sub3_sub3_data = bin1_sub3_data[bin1_sub3_data['X3_sub_sub_bin'] == 'Bin1_Sub3_Sub3']
defect_rate_sub3_sub3 = bin1_sub3_sub3_data['Defect'].mean()

print(f"Bin1_Sub3_Sub3의 불량률: {defect_rate_sub3_sub3:.2%}")
```
