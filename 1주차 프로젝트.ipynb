```python
import pandas as pd
# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df = pd.read_csv('1Ï£º_Ïã§ÏäµÎç∞Ïù¥ÌÑ∞.csv')
columns_to_check = ['X1','X2','X3','X5','X6','X7','X8','X9','X10','X11','X12','X14','X15','X16','X17']

# Í∞Å Î≥ÄÏàòÎ≥Ñ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)  # 1ÏÇ¨Î∂ÑÏúÑÏàò
    Q3 = df[column].quantile(0.75)  # 3ÏÇ¨Î∂ÑÏúÑÏàò
    IQR = Q3 - Q1  # IQR Í≥ÑÏÇ∞
    lower_bound = Q1 - 1.5 * IQR  # Ïù¥ÏÉÅÏπò ÌïòÌïúÏÑ†
    upper_bound = Q3 + 1.5 * IQR  # Ïù¥ÏÉÅÏπò ÏÉÅÌïúÏÑ†

    # Ïù¥ÏÉÅÏπò Ï†úÍ±∞
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ ÌõÑÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†•
print(df)
```

# 4.Ïù¥ÏÉÅÏπò Ï†úÍ±∞

```python
bounds = {}

for column in data.columns:
    lower_bound = data[column].quantile(0.01)
    upper_bound = data[column].quantile(0.99)
    bounds[column] = (lower_bound, upper_bound)

# Î™®Îì† Ïª¨ÎüºÏóê ÎåÄÌïú Ï°∞Í±¥ÏùÑ ÎèôÏãúÏóê Ï†ÅÏö©ÌïòÏó¨ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
for column in data.columns:
    data = data[(data[column] >= bounds[column][0]) & (data[column] <= bounds[column][1])]

# Ïù¥ÏÉÅÏπòÍ∞Ä Ï†úÍ±∞Îêú Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
data.columns
```

# 5.t-test Í≤ÄÏ†ï

```python
import pandas as pd
import numpy as np
import scipy
## T-test Í≤ÄÏ†ï
t_test = []

for idx, col in enumerate(data.columns):
    t = scipy.stats.ttest_ind(data[data['Y']==0][col],
                              data[data['Y']==1][col])
    t_test.append([col, t[0], t[1]])

t_test_data = pd.DataFrame(t_test, columns=['col', 't', 'p-value'])

## P-valueÍ∞Ä 0.05Î≥¥Îã§ ÏûëÏùÄ Î≥ÄÏàòÎßå Ï∂îÏ∂ú
t_test_data = t_test_data[t_test_data['p-value']<0.05]
t_test_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
import pandas as pd
import matplotlib.pyplot as plt
# ÌäπÏÑ±Í≥º ÌÉÄÍ≤ü Î≥ÄÏàò Î∂ÑÎ¶¨
X = data.drop('Y', axis=1)
y = data['Y']

# Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# XGBoost Î™®Îç∏ ÌõàÎ†®
model = xgb.XGBClassifier(objective='binary:logistic', random_state=42)
model.fit(X_train, y_train)

# Î≥ÄÏàò Ï§ëÏöîÎèÑ Ï∂îÏ∂ú
feature_importances = model.feature_importances_
```

```python
import matplotlib.pyplot as plt
import pandas as pd
features = X_train.columns

# Í∞ÄÏ†ï: feature_importancesÎäî Ïù¥ÎØ∏ Í≥ÑÏÇ∞Îêú ÌîºÏ≤ò Ï§ëÏöîÎèÑ Î∞∞Ïó¥
# features = X_train.columns # X_trainÏù¥ pandas DataFrameÏùº Í≤ΩÏö∞
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # yÏ∂ï ÏàúÏÑú Îí§ÏßëÍ∏∞

# xÏ∂ï ÎààÍ∏à Í∞ÑÍ≤© ÏÑ§Ï†ï
plt.xticks(np.arange(min(importance_df['Importance']), max(importance_df['Importance']) + 0.1, 0.1))

plt.show()
```

```python
# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú 'x3' Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú 10Í∞úÏùò Íµ¨Í∞ÑÏúºÎ°ú Î∂ÑÌï†Ìï©ÎãàÎã§. Ïù¥ Îïå Íµ¨Í∞ÑÏùò Í≤ΩÍ≥ÑÎäî 'cut_points'Ïóê Ï†ÄÏû•Îê©ÎãàÎã§.
cut_points = pd.cut(data['X3'], bins=10)

# Í∞Å Íµ¨Í∞ÑÎ≥ÑÎ°ú Î∂àÎüâ Ïó¨Î∂ÄÎ•º ÌôïÏù∏ÌïòÏó¨ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
# 'groupby'Î•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞Å Íµ¨Í∞ÑÏóê ÎåÄÌïú Î∂àÎüâ Ïó¨Î∂ÄÏùò ÌèâÍ∑†ÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
defect_rate_by_bin = data.groupby(cut_points)['Y'].mean()

# Í∞Å Íµ¨Í∞ÑÎ≥Ñ Î∂àÎüâÎ•† Ï∂úÎ†•
print(defect_rate_by_bin)

# Í∞Å Íµ¨Í∞ÑÏóêÏÑúÏùò Î∂àÎüâ Î∞úÏÉù ÎπÑÏú®ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Î∂àÎüâ Ïó¨Î∂ÄÎ•º ÌåêÎã®Ìï©ÎãàÎã§.
threshold = 0.1  # Î∂àÎüâ Î∞úÏÉù ÎπÑÏú®Ïùò ÏûÑÍ≥ÑÍ∞í
defective_prediction = defect_rate_by_bin > threshold

# Í∞Å Íµ¨Í∞ÑÏóêÏÑúÏùò Î∂àÎüâ Ïó¨Î∂ÄÎ•º Ï∂úÎ†•Ìï©ÎãàÎã§.
print(defective_prediction)
```

```python
# X3Î•º 10Í∞úÏùò Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÍ∏∞
data['X3_bin'] = pd.cut(data['X3'], bins=10, labels=[f'Bin{i+1}' for i in range(10)])

# Í∞Å Íµ¨Í∞ÑÏóêÏÑúÏùò Î∂àÎüâÎ•† Í≥ÑÏÇ∞
data['Defect'] = data['Y'] == 1  # Î∂àÎüâÌíà Ïó¨Î∂ÄÎ•º ÎÇòÌÉÄÎÇ¥Îäî ÏÉàÎ°úÏö¥ Ïó¥ Ï∂îÍ∞Ä
grouped_data = data.groupby('X3_bin')['Defect'].mean().reset_index()
grouped_data.columns = ['X3_Bin', 'Defect_Rate']

# ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 6))
sns.barplot(x='X3_Bin', y='Defect_Rate', data=grouped_data, palette='coolwarm')
plt.title('Defect Rate by X3 Bins')
plt.xlabel('X3 Bins')
plt.ylabel('Defect Rate')
plt.show()
```
